3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
def plot_ellipse(a, b=0, hyper=False, name='График'):
    
    plt.axis('equal')
    
    def _plot_circle(x, y, name, hyper=False):
        
        plt_color = 'blue'
        plt.grid()
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title(name)
        
        if not hyper:
            plt.plot(x, y, color=plt_color)
            plt.plot(x, -y, color=plt_color)
        else:
            plt.plot(x, y, color=plt_color)
            plt.plot(x, -y, color=plt_color)
            plt.plot(-x, y, color=plt_color)
            plt.plot(-x, -y, color=plt_color)
        
        return plt.show()
    
    if not hyper:
        x=np.array(np.linspace(-a, a, 2*a*100+1))
        y=np.sqrt(b**2*(1-(x**2/a**2)))  
        return _plot_circle(x, y, name)
    else:
        x=np.array(np.linspace(a, 2*a, a*100+1))
        y=np.sqrt(b**2*((x**2/a**2)-1))
        return _plot_circle(x, y, name, hyper=True)

    
plot_ellipse(a=5, b=5, hyper=False, name='Круг')
plot_ellipse(a=3, b=6, hyper=False, name='Эллипс')
plot_ellipse(a=1, b=2, hyper=True, name='Гипербола')
```

y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
x = np.linspace(0, 10, 1001)
def f_cos(x, k=1, a=0, b=0):
    return k*np.cos(x-a)+b

for k_itr in range(1, 5):
    y = f_cos(x, k=k_itr)
    plt.plot(x, y)
plt.grid()
plt.legend(labels=['k=1', 'k=2', 'k=3', 'k=4'])
plt.title('y(x) = k∙cos(x – a) при различных k')
plt.show()

for a_itr in range(0, 4):
    y = f_cos(x, a=a_itr)
    plt.plot(x, y)
plt.grid()
plt.legend(labels=['a=0', 'a=1', 'a=2', 'a=3'])
plt.title('y(x) = k∙cos(x – a) при различных a')
plt.show()
    
for b_itr in range(0, 4):
    y = f_cos(x, b=b_itr)
    plt.plot(x, y)
plt.grid()
plt.legend(labels=['b=0', 'b=1', 'b=2', 'b=3']);
plt.title('y(x) = k∙cos(x – a) при различных b')
plt.show()
```

3. Задание (в программе)
Напишите код, который будет переводить полярные координаты в декартовы.

Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.

```python
def polar_to_dec(r, phi):
    return r*np.cos(phi), r*np.sin(phi)

def plot_polar(phi, r):
    plt.polar(phi, r)
    return plt.show()

def decar_to_polar(x, y):
    r = np.sqrt(x^2 + y^2)
    return r, (np.arctan(0) if x==0 else np.arctan(y/x))


def segm_in_dec_on_polar(x1, y1, x2, y2):
    r1, phi_1 = decar_to_polar(x1, y1)
    r2, phi_2 = decar_to_polar(x2, y2)
    phi_seg = [phi_1, phi_2]
    r_seg = [r1, r2]
    return plot_polar(phi_seg, r_seg)

def segm_in_polar(phi, p, a):
    r = p/(np.cos(phi-a))
    return plot_polar(phi, r)
    

print(polar_to_dec(25, 0.5))
PHI = np.linspace(0, np.pi*2, 3600)
plot_polar(PHI, [25] * len(PHI));
segm_in_dec_on_polar(0, 1, 6, 8)
phi_cord = np.array([-2, 2])
segm_in_polar(phi_cord, p=3, a=1)
```

4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1
Решите систему уравнений и неравенств:
y = x2 – 1
exp(x) + x∙(1 – y) > 1

```python
from scipy.optimize import fsolve

def f1(x):
    return (np.e**x+x-1)/x

def f2(x):
    return x**2-1

def f(x):
    return f1(x) - f2(x)

x = np.linspace(-3, 6, 1500)
plt.plot(x, f1(x))
plt.plot(x, f2(x))
plt.grid()

x1 = fsolve(f, -2)
x2 = fsolve(f, 2)
x3 = fsolve(f, 4)
x0 = [0]
print(x1, x0, x2, x3)
```

```python
x_1 = np.linspace(-3, x1, 150)
x_2 = np.linspace(x1, x2, 150)
x_3 = np.linspace(x2, x3, 150)
x_4 = np.linspace(x3, 6, 150)

plt.plot(x_1, f(x_1), color='white')
plt.plot(x_2, f(x_2), color='blue')
plt.plot(x_3, f(x_3), color='white')
plt.plot(x_4, f(x_4), color='blue')
plt.grid()
f"x принадлежит ({round(x1[0], 4):3}; 0), (0, {round(x2[0], 4):3}), ({round(x3[0], 4):3}; INF)"
```
